```python
hygdf = pd.read_csv(data_v3_path)
hygdf.info()
```

```python
selected_columns = ['id','hd','hr','proper','dist','mag','absmag','x','y','z']
selected_columns
```

```python
fillna_values = {'hd': 0, 'hr': 0, 'proper': '', 'dist': 0, 'mag': 0, 'absmag': 0, 'x': 0, 'y': 0, 'z': 0}
hygdf[selected_columns].mask(hygdf['dist'] == 100000.0, ).dropna(how='all').fillna(value=fillna_values).set_index('id')
```

```python
hygdf[selected_columns].mask(hygdf['dist'] == 100000.0, ).dropna(how='all').fillna('').set_index('id')
```

## Save CSV file with only required data to minimize its size

```python
# Removing all the stars with any missing data to form a miniature dataset
hygdf_min = hygdf[selected_columns].mask(hygdf['dist'] == 100000.0, ).dropna(how='any').set_index('id')
hygdf_min
```

```python
hygdf_min.sort_values(by=['mag','dist'],ascending = False, na_position ='last', inplace=True)
hygdf_min.to_csv('./HYG-Database/hygdata_v3_min.csv')
```
